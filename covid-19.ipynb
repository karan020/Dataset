Step1. import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=(10,5)
```

Step2.we will load the DataSet

```python
df=pd.read_csv("COVID-19 Dataset.csv")
print("DataSet Shape")
print(df.shape)
df.head(10)
```

Step3. Dataset Information

```python
df.info()
```

Step4.Summary Statics

```python
df.describe()
```

Step5. Missing value Anlysis

```python
missing_values=df.isnull().sum()

missing_values
```

Step6.visualize missing values

```python
missing_values.plot(kind="bar")
plt.title("Missing value bar")
plt.xlabel("column")
plt.ylabel("count")

plt.show()
```

Step7. Remove row with more than 2 rows

```python
initial_rows=len(df)

df=df[
    df.isnull().sum(axis=1)<=2
]

rows_removed=(
    initial_rows - len(df)
)

print(
    "rows removed",
    rows_removed
)
```

Step8.check Dataset type

```python
df.dtypes
```

Step9. Standardize text Columns

```python
if "Country/Region" in df.columns:

    df["Country/region"]=(
        df["Country/Region"]
        .astype(str)
        .str.upper()
        .str.strip()
    )

df.head()
```

Step10.Duplicate Analysis

```python
duplicates=(
    df.duplicated().sum()
    )
print(
    "Duplicate Rows"
)
```

Step11. Remove Duplicates

```python
df=df.drop_duplicates()

print(
    "Duplicates Remainig",
     df.duplicated().sum()
)
```

Step 12.Create Death Rate

```python
df["Death_Rate"]=(
    df["Deaths"]
    /
    df["Confirmed"]
)*100

df.head()
```

Step3.Create Recovery Rate

```python
df["Recovery_Rate"]=(
    df["Recovered"]
    /
    df["Confirmed"]
)*100

df.head()
```

Step14.Create Active Rate

```python
df["Active_Rate"]=(
    df["Active"]
    /df["Confirmed"]
)

df.head()
```

Step15.Histogram for Comfirmed Cases

```python
plt.hist(
    df["Confirmed"],
    bins=20
)

plt.title("COnfirmed Cases Obtained")

plt.xlabel("Confirmed Cases")

plt.ylabel("Frequency")

plt.show()
```

Step16.Top 10 Countries By Confirmed Cases

```python
top10=(
    df.sort_values(
        by="Confirmed",
        ascending=False
    )
    .head(10)
)

plt.bar(
    top10["Country/Region"],
    top10["Confirmed"]
)

plt.title("Top 10 Countries With Confirmed Cases")

plt.xticks(rotation=90)

plt.show()
```

Step17.Top 10 Countries By Death

```python
top10_deaths=(
     df.sort_values(
        by="Deaths",
        ascending=False
    )
    .head(10)
)

plt.bar(
    top10_deaths["Country/Region"],
    top10_deaths["Deaths"]
)

plt.title("Top 10 Countries by Death")

plt.xticks(rotation=90)

plt.show()
```

Step18.Confirmed Vs Deaths

```python
plt.scatter(
    df["Confirmed"],
    df["Deaths"]
)

plt.title("Confirmed vs Deaths")

plt.xlabel("Confirmed")
plt.ylabel("Deaths")

plt.show()
```

Step19. WHO Region Distribution

```python
region_cases=(
    df.groupby("WHO Region")
    ["Confirmed"].sum()
)

plt.bar(
    region_cases.index,
    region_cases.values
)

plt.title("Cases by WHO Region")

plt.xticks(rotation=45)

plt.show()
```

Step20.Recovery Rate Distribution

```python
plt.hist(
    df["Recovered"],
    bins=20
)

plt.title(" Recovery Rate Distribution")

plt.xlabel("Recovery Rate(%)")

plt.ylabel("Frequency")

plt.show()
```

Step 21. Death Rate Distibution

```python
plt.hist(
    df["Death_Rate"],
    bins=20
)
plt.title("Death Rate Distribution")

plt.xlabel("Death Rate(%)")
plt.ylabel("Frequency")

plt.show()
```

Step 22. Box Plot

```python
plt.boxplot(
    [
        df["Confirmed"],
        df["Deaths"],
        df["Recovered"]
    ],
    label=[
        "Confirmed",
        "Death",
        "Recovered"
    ]
)

plt.title("Covid Outlier Analysis")

plt.show()
```

Step 23.Correltion Matrix

```python
corr = df[
    [
        "Confirmed",
        "Deaths",
        "Recovered",
        "Active",
        "New cases",
        "New deaths",
        "New recovered"
    ]
].corr()
plt.imshow(
    corr,
    cmap="coolwarm"
)

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Matrix")

plt.show()
```

Step24.Save clean DataSet

```python
df.to_csv(
    "COVID-19 Dataset.csv",
    index=False
)

print(
    "Cleaned file saved succesfully"
)
```
